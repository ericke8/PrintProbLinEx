```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
img = cv.imread("anon_blank_blank_00height_amaske-6912_page1r.tif", 0)
ret,bin_img = cv.threshold(img,127,255,cv.THRESH_BINARY)

plt.figure(figsize=(5,10))
plt.imshow(img, cmap="gray")

plt.figure(figsize=(5,10))
plt.imshow(bin_img, 'gray') 
print(img.shape)

plt.show()
```

```python
print("Number of Rows: " +  str(bin_img.shape[0]))
print("Number of Columns: " + str(bin_img.shape[1]))

# Width of Expansion
expandWidth = 50;

# Create new image
# grow_img = [[255]*bin_img.shape[1]]*bin_img.shape[0]
ret,grow_img = cv.threshold(img,127,255,cv.THRESH_BINARY)
print("grow_img shape: " + str(img.shape))


# if grow_img[0][0] == 255:
#     print("HELLO?")

bin_img[20][40] = 0

for row in range(bin_img.shape[0]):
    for col in range(bin_img.shape[1]):
        # Look at Pixels to the Left
        for left in range(1,expandWidth+1):
            # Don't exand if pixel is not black or if reaching out of bounds
            if bin_img[row][col] == 255 or col - left < 0:
              break
            else:
              # Expand left
              grow_img[row][col - left] = 0
        # Look at Pixels to the Right    
        for right in range(1,expandWidth+1):
            # Don't exand if pixel is not black or if reaching out of bounds
            if bin_img[row][col] == 255 or col + right > bin_img.shape[1]-1:
              break
            else:
                # Expand right
               grow_img[row][col + right] = 0
              
plt.figure(figsize=(5,10))
plt.imshow(grow_img, 'gray') 
plt.figure(figsize=(5,10))
plt.imshow(bin_img, 'gray') 
                        

plt.show()
```
