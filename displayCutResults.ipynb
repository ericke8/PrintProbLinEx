```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
fileName = "anon_blank_blank_00height_amaske-6913_page1r"
```

```python
img = cv.imread(fileName + ".tif", 0)
sol_rows = pd.read_csv(fileName + "_splits.csv", header=None, names=["row"])
```

```python
fig, ax = plt.subplots(figsize= (20,20))
ax.imshow(img, cmap="gray")
for vl in sol_rows["row"]:
    ax.axhline(y=vl, color="red")
plt.show()
```

```python
split = sol_rows["row"]
for row in range(0,split.shape[0] - 1):
    cutLine = img[split[row]:split[row+1],:]
    plt.imshow(cutLine, cmap="gray")
    plt.show()
```
